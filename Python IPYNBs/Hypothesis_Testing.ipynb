```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr, linregress
# Load the dataset
file_path = "../DataSet/BSc1DADataSetFinal.xlsx"  # Update path if needed
df = pd.read_excel(file_path, header=2)
# Convert columns to numeric, coercing errors to NaN
df['Practicals'] = pd.to_numeric(df['Practicals'], errors='coerce')
df['GrandTotal'] = pd.to_numeric(df['GrandTotal'], errors='coerce')
# Drop rows with missing values in either column
data = df[['Practicals', 'GrandTotal']].dropna()

# Pearson correlation
pearson_corr, pearson_p = pearsonr(data['Practicals'], data['GrandTotal'])
print(f"Pearson correlation: {pearson_corr:.3f}, p-value: {pearson_p:.4g}")
# Spearman correlation
spearman_corr, spearman_p = spearmanr(data['Practicals'], data['GrandTotal'])
print(f"Spearman correlation: {spearman_corr:.3f}, p-value: {spearman_p:.4g}")
# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(data['Practicals'], data['GrandTotal'])
print(f"Linear regression: GrandTotal = {slope:.2f}*Practicals + {intercept:.2f}")
print(f"R-squared: {r_value**2:.3f}, p-value: {p_value:.4g}")
# Hypothesis test interpretation
alpha = 0.05
if pearson_p < alpha:
    print("Reject Hâ‚€: There is a significant positive correlation between practical scores and overall performance.")
else:
    print("Fail to reject Hâ‚€: No significant correlation found between practical scores and overall performance.")
# Optional: Visualize regression (scatter + line)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(data['Practicals'], data['GrandTotal'], alpha=0.5, label='Data')
plt.plot(data['Practicals'], intercept + slope*data['Practicals'], color='red', label='Regression line')
plt.xlabel('Practical Marks')
plt.ylabel('Grand Total')
plt.title('Linear Regression: GrandTotal ~ Practicals')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.stats import f_oneway

# Load the Excel file
df = pd.read_excel("../DataSet/BSc1DADataSetFinal.xlsx", sheet_name="Sheet1")

# Subject combinations from your processed data
subject_combos = [
    ("Mathematics", "Physics", "Statistics"),
    ("Chemistry", "Mathematics", "Physics"),
    ("Geography", "Mathematics", "Physics"),
    ("Botany", "Chemistry", "Zoology"),
    ("Computer Science", "Mathematics", "Physics"),
    ("Defence Studies", "Mathematics", "Physics"),
    ("Computer Science", "Mathematics", "Psychology"),
    ("Geology", "Mathematics", "Physics"),
    ("Anthropology", "Botany", "Zoology"),
    ("Biochemistry", "Chemistry", "Zoology"),
    ("Botany", "Defence Studies", "Zoology"),
    ("Biochemistry", "Botany", "Chemistry"),
]

# Generate names for combinations
combo_names = [
    " + ".join(combo) for combo in subject_combos
]

# Create a DataFrame
combo_df = pd.DataFrame(subject_combos, columns=["Sub1", "Sub2", "Sub3"])
combo_df["ComboName"] = combo_names

# Assign each student to a combination if they have all three subjects
def get_combo(row):
    for i, combo in combo_df.iterrows():
        if pd.notna(row.get(combo["Sub1"])) and pd.notna(row.get(combo["Sub2"])) and pd.notna(row.get(combo["Sub3"])):
            return combo["ComboName"]
    return None

# Apply to each row
df["Combo"] = df.apply(get_combo, axis=1)

# Calculate total marks for each student (sum of subjects they have)
subject_list = list(set(combo_df["Sub1"]) | set(combo_df["Sub2"]) | set(combo_df["Sub3"]))
df["TotalMarks"] = df[subject_list].sum(axis=1, skipna=True)

# Group marks by combinations
grouped = df.dropna(subset=["Combo", "TotalMarks"]).groupby("Combo")["TotalMarks"].apply(list)

# Perform ANOVA
anova_result = f_oneway(*grouped)

# Show result
print("=== ANOVA Result ===")
print(f"F-Statistic: {anova_result.statistic}")
print(f"P-Value: {anova_result.pvalue}")

# Optional: print group means
print("\nGroup-wise Averages:")
for combo, marks in grouped.items():
    print(f"{combo}: {sum(marks)/len(marks):.2f}")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load Excel with correct header
df = pd.read_excel("../DataSet/BSc1DADataSetFinal.xlsx", header=2)

# Use EnrolNo as identifier
df['EnrolNo'] = df['EnrolNo']

# Define actual subject groups using exact column names
physics_cols = ['Mechanics and Special Theory of Relativity S.T.R', 'Thermal Phyics', 'Electronics']
maths_cols = ['Geometry ', 'Real Analysis', 'Ordinary Differential Equations O.D.E.']
stats_cols = ['Probability Theory ', 'Statistical Methods']

# Compute total marks for each subject group
df['Physics_Total'] = df[physics_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Mathematics_Total'] = df[maths_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Statistics_Total'] = df[stats_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)

# Combine into one DataFrame
subject_totals = df[['EnrolNo', 'Physics_Total', 'Mathematics_Total', 'Statistics_Total']].copy()

# Convert to long format for ANOVA
df_long = pd.melt(subject_totals, id_vars='EnrolNo', 
                  value_vars=['Physics_Total', 'Mathematics_Total', 'Statistics_Total'],
                  var_name='Subject', value_name='TotalMarks')

# Clean subject names
df_long['Subject'] = df_long['Subject'].str.replace('_Total', '')

# Drop missing values
df_long.dropna(subset=['TotalMarks'], inplace=True)

# Perform one-way ANOVA
model = ols('TotalMarks ~ C(Subject)', data=df_long).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print results
print("ANOVA Results:\n")
print(anova_table)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file from the correct row
df = pd.read_excel("../DataSet/BSc1DADataSetFinal.xlsx", header=2)

# Select actual subject columns based on correct names
physics_cols = ['Mechanics and Special Theory of Relativity S.T.R', 'Thermal Phyics', 'Electronics']
maths_cols = ['Geometry ', 'Real Analysis', 'Ordinary Differential Equations O.D.E.']
stats_cols = ['Probability Theory ', 'Statistical Methods']

# Compute total marks per subject group
df['Physics_Total'] = df[physics_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Mathematics_Total'] = df[maths_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Statistics_Total'] = df[stats_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)

# Create combined DataFrame
subject_totals = df[['EnrolNo', 'Physics_Total', 'Mathematics_Total', 'Statistics_Total']].copy()

# Melt to long format
df_long = pd.melt(subject_totals, id_vars='EnrolNo',
                  value_vars=['Physics_Total', 'Mathematics_Total', 'Statistics_Total'],
                  var_name='Subject', value_name='TotalMarks')

# Clean names and drop missing values
df_long['Subject'] = df_long['Subject'].str.replace('_Total', '')
df_long.dropna(subset=['TotalMarks'], inplace=True)

# ðŸ“Œ Print Hypotheses
print("Null Hypothesis (H0): All subjects have the same average total marks.")
print("Alternative Hypothesis (H1): At least one subject has a different average total marks.\n")

# ðŸ“Š Perform ANOVA
model = ols('TotalMarks ~ C(Subject)', data=df_long).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display results
print("ANOVA Results:\n")
print(anova_table)

# ðŸ“ˆ Plot boxplot for subject-wise comparison
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_long, x='Subject', y='TotalMarks', palette='Set2')
plt.title('Total Marks Distribution Across Subjects')
plt.ylabel('Total Marks')
plt.xlabel('Subject')
plt.grid(True)
plt.tight_layout()
plt.show()

# ðŸ“‰ Print group means
print("\nMean Total Marks per Subject:")
print(df_long.groupby('Subject')['TotalMarks'].mean())
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file (adjust path if needed)
df = pd.read_excel("../DataSet/BSc1DADataSetFinal.xlsx", header=2)

# Exclude non-subject columns
excluded_cols = ['RollNo', 'EnrolNo', 'Name of the Student', 'Result', 'Division']
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
subject_cols = [col for col in numeric_cols if col not in excluded_cols]

# Group subject columns by prefix
subject_groups = {}
for col in subject_cols:
    key = col.split()[0]
    subject_groups.setdefault(key, []).append(col)

# Keep only subjects with at least 2 columns (e.g., 2 papers per subject)
valid_subject_groups = {
    subject: [col for col in cols if col in df.columns]
    for subject, cols in subject_groups.items()
    if len([col for col in cols if col in df.columns]) >= 2
}

# Proceed only if we have at least 2 subject groups
if len(valid_subject_groups) >= 2:
    valid_total_cols = []
    for subject, cols in valid_subject_groups.items():
        df[f'{subject}_Total'] = df[cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)
        valid_total_cols.append(f'{subject}_Total')

    # Melt into long format for ANOVA
    subject_totals_valid = df[['EnrolNo'] + valid_total_cols].copy()
    df_long = pd.melt(subject_totals_valid, id_vars='EnrolNo',
                      value_vars=valid_total_cols,
                      var_name='Subject', value_name='TotalMarks')
    df_long['Subject'] = df_long['Subject'].str.replace('_Total', '')
    df_long.dropna(subset=['TotalMarks'], inplace=True)

    # Run ANOVA
    model = ols('TotalMarks ~ C(Subject)', data=df_long).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # Print hypotheses
    print("Null Hypothesis (Hâ‚€): All subjects have the same average total marks.")
    print("Alternative Hypothesis (Hâ‚): At least one subject has a significantly different average total mark.")
    print("\nANOVA Table:\n", anova_table)

    # Plot boxplot for visualization
    plt.figure(figsize=(14, 6))
    sns.boxplot(data=df_long, x='Subject', y='TotalMarks', palette='Set2')
    plt.title('Distribution of Total Marks Across Subjects')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

else:
    print("Not enough valid subject groups with multiple components to perform ANOVA.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "Merged_Yearwise_StudentDetails.xlsx"
df = pd.read_excel(file_path, header=1)

# Filter student
student_id = "U2112063"
student_row = df[df['EnrolNo'] == student_id]

# Extract subject total columns
total_columns = [col for col in student_row.columns if 'Total' in col]
year1_totals = {col: student_row[col].values[0] for col in total_columns if col.endswith('1Total')}
year2_totals = {col: student_row[col].values[0] for col in total_columns if not col.endswith('1Total') and not col.startswith('Part') and not col.startswith('Grand')}

# Prepare labels and scores, filtering out 0s
labels, year1_scores, year2_scores = [], [], []
for subject in sorted(set(year1_totals) | set(year2_totals)):
    mark1 = pd.to_numeric(year1_totals.get(subject, 0), errors='coerce')
    mark2 = pd.to_numeric(year2_totals.get(subject, 0), errors='coerce')
    if (pd.notna(mark1) and mark1 > 0) or (pd.notna(mark2) and mark2 > 0):
        label = subject.replace('1Total', '').replace('Total', '').strip()
        labels.append(label)
        year1_scores.append(mark1)
        year2_scores.append(mark2)

# Plotting with adjacent bars per subject
bar_width = 0.4
x = range(len(labels))

plt.figure(figsize=(12, 6))
plt.bar([i for i in x], year1_scores, width=bar_width, label='Year 1', color='skyblue')
plt.bar([i + bar_width for i in x], year2_scores, width=bar_width, label='Year 2', color='salmon')

# Adjust x-axis ticks to center between Year 1 and Year 2 bars
plt.xticks([i + bar_width / 2 for i in x], labels, rotation=45, ha='right')

plt.xlabel("Subjects")
plt.ylabel("Total Marks")
plt.title(f"Subject-wise Total Marks Comparison for {student_id}")
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "Merged_Yearwise_StudentDetails.xlsx"
df = pd.read_excel(file_path, header=1)

# Filter student
student_id = "U2112063"
student_row = df[df['EnrolNo'] == student_id]

# Extract total columns
total_columns = [col for col in student_row.columns if 'Total' in col and not col.startswith('Part') and not col.startswith('Grand')]

# Normalize subject names (remove year and 'Total')
subject_scores = {}
for col in total_columns:
    subject = col.replace('1Total', '').replace('Total', '').strip()
    year = 'Year 1' if '1Total' in col else 'Year 2'
    mark = pd.to_numeric(student_row[col].values[0], errors='coerce')
    
    if pd.notna(mark) and mark > 0:
        if subject not in subject_scores:
            subject_scores[subject] = {'Year 1': 0, 'Year 2': 0}
        subject_scores[subject][year] = mark

# Prepare data for plotting
subjects = list(subject_scores.keys())
year1_scores = [subject_scores[sub]['Year 1'] for sub in subjects]
year2_scores = [subject_scores[sub]['Year 2'] for sub in subjects]

# Plotting
bar_width = 0.35
x = range(len(subjects))

plt.figure(figsize=(12, 6))
plt.bar([i - bar_width/2 for i in x], year1_scores, width=bar_width, label='Year 1', color='skyblue')
plt.bar([i + bar_width/2 for i in x], year2_scores, width=bar_width, label='Year 2', color='salmon')

plt.xticks(ticks=x, labels=subjects, rotation=45, ha='right')
plt.xlabel("Subjects")
plt.ylabel("Total Marks")
plt.title(f"Subject-wise Total Marks Comparison for {student_id}")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
